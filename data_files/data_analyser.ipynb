```python
import csv

#load csv and  return a list of dictionaries
def load_csv(file_path):

    data = []
    with open(file_path, 'r', newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return data


#filter predictions based on sentiment value
def filter_predictions_from_csv(file_path, sentiment_value=0):
   
    predicted_data = load_csv(file_path)
    filtered_predictions = [entry for entry in predicted_data if int(entry['sentiment']) == sentiment_value]
    
    return filtered_predictions



predicted_file_path = '../submission.csv'


#sentiment value 0 = negative, 1 = positive
negative_predictions = filter_predictions_from_csv(predicted_file_path, sentiment_value=0)
positive_predictions = filter_predictions_from_csv(predicted_file_path, sentiment_value=1)


#print negative predictions
negative_predictions
```
